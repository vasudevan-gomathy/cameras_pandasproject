# Importing necessary packages

```python
import pandas as pd
import numpy as np
```

# Creating Dataframe
Created Dataframe "camera_data" using CSV file

```python
camera_data = pd.read_csv("/content/Camera.csv")
camera_data.head()
```

# Finding NULL values
Finding out the percentage of blank values in each column.

```python
# percent_missing = camera_data.isnull()
camera_data.isnull().sum() * 100 / len(camera_data)
```

# Statistical summary of data

```python
camera_data.describe()
```

```python
camera_data.info()
```

# Replacing all the Blank values with the column median.

```python
# Identify the columns to fill (excluding 'model')
cols_to_fill = camera_data.columns.difference(['Model'])

# Replace NaN with median values in the selected columns
camera_data[cols_to_fill] = camera_data[cols_to_fill].apply(lambda x: x.fillna(x.median()))
```

```python
camera_data.isnull().sum()
```

# Adding new field "discounted_price"

Discounted_price is the discount of 5% in the Price column.

```python
camera_data['discounted_price'] = camera_data['Price'] * .95
```

```python
camera_data.head()
```

# Drop the columns Zoom Tele & Macro Focus range

```python
camera_data = camera_data.drop(['Zoom tele (T)', 'Macro focus range'], axis = 1)
```

# Replacing the Model Name Agfa ePhoto CL50 with Agfa ePhoto CL250

```python
camera_data['Model'] = camera_data['Model'].replace('Agfa ePhoto CL50', 'Agfa ePhoto CL250')
```

# Renaming the column name from Release Date to Release Year

```python
camera_data = camera_data.rename(columns={'Release date': 'Release Year'})
```

# Finding the most expensive Camera

```python
# Finding the row with the highest price
max_price_row = camera_data[camera_data['Price'] == camera_data['Price'].max()]

# Display the model and its corresponding price
result = max_price_row[['Model', 'Price']]
print(result)
```

Here all 1) Canon EOS-1Ds, 2) Canon EOS-1Ds Mark II, 3) Canon EOS-1Ds Mark III shares the same price and hence all the above models are most expensive

# Finding camera having the least weight

```python
# Finding the row with least weight
min_weight = camera_data[camera_data['Weight (inc. batteries)'] == camera_data['Weight (inc. batteries)'].min()]

# Displaying the model and its corresponding weight
result = min_weight[['Model', 'Weight (inc. batteries)']]
print(result)
```

The model "Casio Exilim EX-S20" has the least weight of 100 grams

# Grouping the data on the basis of their release year.

```python
# Group by 'Release Year' and aggregate the models
grouped_models = camera_data.groupby('Release Year')['Model'].apply(', '.join).reset_index()

# Display the grouped models
print(grouped_models)
```

# Extracting the Name, Storage Include, Price, Disounted_Price & Dimensions columns

```python
extracted_data1 = camera_data[["Model", "Storage included", "Price", "discounted_price", "Dimensions"]]
extracted_data1
```

# Extract the records for the cameras released in the year 2005 & 2006

```python
# Extracting records for cameras released in 2005 and 2006
filtered_data = camera_data[camera_data['Release Year'].isin([2005, 2006])]

# Display the filtered data
filtered_data.head()
```

# Finding out 2007s expensive & Cheapest Camera.

```python
# Filtering for cameras released in 2007
cameras_2007 = camera_data[camera_data['Release Year'] == 2007]

# Finding the most expensive camera
most_expensive = cameras_2007.loc[cameras_2007['Price'].idxmax()]

# Finding the cheapest camera
cheapest = cameras_2007.loc[cameras_2007['Price'].idxmin()]

# Displaying the results
print("Most Expensive Camera in 2007:")
print(most_expensive)

print("\nCheapest Camera in 2007:")
print(cheapest)
```

# Finding out the year where maximum number of models is released?

```python
# Counting the number of models released each year
release_counts = camera_data['Release Year'].value_counts()

# Finding the year with the maximum number of models released
max_year = release_counts.idxmax()
max_count = release_counts.max()

# Displaying the results
print(f"The year with the maximum number of models released is {max_year} with {max_count} models.")
```
